```python
from databricks.sdk import WorkspaceClient
```

```python
workspace_client = WorkspaceClient()
```

```python
space_id = dbutils.widgets.get("genie-space-id")
warehouse_id = dbutils.widgets.get("Warehouse_id")
print(f"{space_id} --- {warehouse_id}")
```

```python
genie_space = workspace_client.genie.get_space(space_id=space_id, include_serialized_space=True)
print(genie_space.serialized_space)
genie_serialized_space = genie_space.serialized_space
```

```python
genie_space_description = genie_space.description
imported_genie_space_title=f"Imported_via_SDK_{genie_space.title}"
print(genie_space_description)
print(imported_genie_space_title)
```

```python
imported_genie_space_description = f"This is an imported Genie space using Genie Export and Import SDK feature. {genie_space_description}"
print(imported_genie_space_description)
```

```python
new_genie_space = workspace_client.genie.create_space(title=imported_genie_space_title, description=imported_genie_space_description, serialized_space=genie_serialized_space, warehouse_id=warehouse_id)
new_genie_space_id = new_genie_space.space_id
print(new_genie_space_id)
```

```python
updated_genie_space_description=f"Here is an updated description. {imported_genie_space_description}"
print(updated_genie_space_description)
updated_genie_space_title = genie_space_title.replace("Imported", "Updated")
print(updated_genie_space_title)
updated_warehouse_id="383cc3b75ffe2072"
```

```python
updated_genie_space = workspace_client.genie.update_space(space_id=new_genie_space_id, title=updated_genie_space_title, description=updated_genie_space_description, warehouse_id=updated_warehouse_id)
print(f"{updated_genie_space.title}\n\n\n")
print(updated_genie_space.description)
```
